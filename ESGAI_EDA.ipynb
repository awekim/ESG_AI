```python
# from openai import OpenAI
# client = OpenAI(api_key="YOUR_API_KEY")

import ollama
import pandas as pd
import re

desiredModel = 'llama3.3'

def AI_detect(transcript):
    prompt = f"""
    Transcript: {transcript}
    **Task** Given a description of a company, determine whether it is AI-related.

    **Criteria for Evaluation**
    AI-related company: The company directly develops AI technologies (e.g., AI models, algorithms, software) or has AI as a core part of its business (e.g., AI-powered solutions, AI research).
    Non-AI company: The company only uses AI as a supplementary tool (e.g., AI for marketing analytics, automation) or has no significant connection to AI.

    **Output Format**
    AI Company Status: (Choose either "AI-related company" or "Non-AI company")
    Explanation: (For AI-related companies, summarize how they utilize AI. For non-AI companies, describe their primary industry and core business activities.)
    """
    # response = client.chat.completions.create(model="gpt-4", messages=[{"role": "system", "content": prompt}])
    # return response.choices[0].message.content
    response = ollama.chat(model=desiredModel, messages=[{"role": "user", "content": prompt}])
    result_text = response.get('message', {}).get('content', 'No response received')
    # response = ollama.chat(model=desiredModel, messages=[{"role": "system", "content": prompt}])
    # print(response)
    # return response['message']['content']

    status_match = re.search(r"AI Company Status:\s*(.*)", result_text)
    explanation_match = re.search(r"Explanation:\s*(.*)", result_text, re.DOTALL)

    status = status_match.group(1).strip() if status_match else "Unknown"
    explanation = explanation_match.group(1).strip() if explanation_match else "No explanation provided"

    return status, explanation

esg_df = pd.read_csv('I:/Data_for_practice/ESG/esg_score_deal_data.csv')
esg_df = esg_df[esg_df['Deal_status'].isin(['Completed', 'Completed Assumed'])]
esg_df = esg_df[['Identifier', 'Year', 'ESG_score', 'Deal_number', 'Deal_type', 'Deal_headline', 'Targer_name', 'Target_overview', 'Target_sector','Target_business_description', 'Target_trade_description']]
esg_df = esg_df[esg_df['Target_overview'].notnull()]
esg_df = esg_df[['Targer_name','Target_overview']].drop_duplicates()
esg_df = esg_df.reset_index(drop=True)
firm_text = esg_df['Target_overview']
firm_text

results = []
for text in firm_text:
    status, explanation = AI_detect(text)
    results.append({
        "Company Description": text,
        "AI Company Status": status,
        "Explanation": explanation
    })
    print(text)

df = pd.DataFrame(results)
df
```
